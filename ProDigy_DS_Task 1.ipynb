 # Import Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Enter Few Population Data

```python
data = {
    'Age': [21, 22, 23, 24, 25, 25, 26, 26, 27, 27, 27, 28, 29, 30, 30, 30, 31, 32, 32, 33, 34, 35],
    'Gender': ['F', 'M', 'F', 'M', 'M', 'M', 'M', 'F', 'M', 'F', 'M', 'M', 'F', 'M', 'M', 'M', 'F', 'F', 'M', 'M', 'F', 'F']
}
```

# Create Data Frame

```python
df = pd.DataFrame(data)
```

# Ploting Age Distribution in Population

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# Ploting Gender Distribution in Population

```python
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(10, 6))
gender_counts.plot(kind='bar', color=['blue', 'orange'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```
